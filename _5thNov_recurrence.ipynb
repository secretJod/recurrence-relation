```python
# 2. Given a rec%rrence relation, solve it %sing the s%bstit%tion method:

# a. T(n) = T(n-1) + c
# The recurrence relation \(T(n) = T(n-1) + c\) can be solved iteratively. Let's expand it:


# T(n) &= T(n-1) + c \\
# = (T(n-2) + c) + c \\
# = T(n-2) + 2c \\
# = T(n-3) + 3c \\
# = \ldots \\
# = T(1) + (n-1)c


# Now, substitute the base case \(T(1)\) with an initial condition, let's say \(T(1) = a\):

# [ T(n) = a + (n-1)c ]

# So, the solution to the recurrence relation \(T(n) = T(n-1) + c\) is \(T(n) = a + (n-1)c\).

```

```python

# b. T(n) = 2T(n/2) + n
# The recurrence relation \(T(n) = 2T\left(\frac{n}{2}\right) + n\) represents the time complexity of an algorithm that divides the problem into two subproblems of half the size and takes linear time \(n\) to combine the results. This type of recurrence relation is common in algorithms using a divide-and-conquer approach.

# Let's analyze the time complexity using the Master Theorem, which is a common method for analyzing divide-and-conquer recurrences of the form \(T(n) = aT\left(\frac{n}{b}\right) + f(n)\).

# In this case, we have:

# - \(a = 2\) (the number of subproblems),
# - \(b = 2\) (the factor by which the problem size is reduced), and
# - \(f(n) = n\) (the time it takes to divide and combine).

# The Master Theorem has the form:

# \[ T(n) = aT\left(\frac{n}{b}\right) + f(n) \]

# The time complexity is determined by comparing \(f(n)\) with \(n^{\log_b a}\). If \(f(n)\) is:

# 1. \(O(n^{\log_b a - \epsilon})\) for some \(\epsilon > 0\), then the time complexity is \(T(n) = \Theta(n^{\log_b a})\).
# 2. \(\Theta(n^{\log_b a})\), then the time complexity is \(T(n) = \Theta(n^{\log_b a} \log n)\).
# 3. \(\Omega(n^{\log_b a + \epsilon})\) for some \(\epsilon > 0\), and \(af\left(\frac{n}{b}\right) \leq kf(n)\) for some constant \(k < 1\) and sufficiently large \(n\), then the time complexity is \(T(n) = \Theta(f(n))\).

# In our case:

# - \(a = 2\), \(b = 2\), \(f(n) = n\),
# - \(n^{\log_b a} = n^{\log_2 2} = n\).

# So, we are in case 2 of the Master Theorem. Therefore, the time complexity is:

# \[ T(n) = \Theta(n \log n) \]

# The time complexity of the algorithm is \(O(n \log n)\).
```

```python

# 3. Given a rec%rrence relation, solve it %sing the rec%rsive tree approach:


# a. T(n) = 2T(n-1) +1



# The recurrence relation \(T(n) = 2T(n-1) + 1\) represents a scenario where a problem of size \(n\) is divided into two subproblems of size \(n-1\), and each division step takes constant time \(1\).

# To analyze the time complexity, let's try to expand the recurrence relation:

# \[
# \begin{align*}
# T(n) &= 2T(n-1) + 1 \\
# &= 2[2T(n-2) + 1] + 1 \\
# &= 2^2 T(n-2) + 2 + 1 \\
# &= 2^3 T(n-3) + 2^2 + 2 + 1 \\
# &= \ldots \\
# &= 2^n T(0) + 2^{n-1} + 2^{n-2} + \ldots + 2 + 1
# \end{align*}
# \]

# Now, if we assume \(T(0)\) is a constant, let's say \(T(0) = a\), then the expression becomes:

# \[ T(n) = a \cdot 2^n + 2^{n-1} + 2^{n-2} + \ldots + 2 + 1 \]

# This recurrence relation has a time complexity of \(O(2^n)\), which is exponential. This is because each level of the recursion tree has twice as many nodes as the previous level.

# However, keep in mind that this type of recurrence relation often results in inefficient algorithms, and it's generally recommended to seek more efficient algorithms for practical purposes.
```
